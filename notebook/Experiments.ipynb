```python
x=data.drop(columns=['income']) #input
y=data['income'] #output
x
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder() # Initialize the LabelEncoder
x['workclass'] = le.fit_transform(x['workclass'])
x['capital-gain'] = le.fit_transform(x['capital-gain'])
x['capital-loss'] = le.fit_transform(x['capital-loss'])
x['hours-per-week'] = le.fit_transform(x['hours-per-week'])
x['educational-num'] = le.fit_transform(x['educational-num'])
x['marital-status'] = le.fit_transform(x['marital-status'])
x['fnlwgt'] = le.fit_transform(x['fnlwgt'])
x['race'] = le.fit_transform(x['race'])
x['relationship'] = le.fit_transform(x['relationship'])
x['gender'] = le.fit_transform(x['gender'])
x['native-country'] = le.fit_transform(x['native-country'])
x['occupation'] = le.fit_transform(x['occupation'])
x['age'] = le.fit_transform(x['age'])
x 
```

```python
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier

# Split the data (X and y are already defined and preprocessed in previous cells)
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
models = {
    "LogisticRegression": LogisticRegression(),
    "RandomForest": RandomForestClassifier(),
}

results = {}

for name, model in models.items():
    pipe = Pipeline([
        ('model', model)
    ])

    pipe.fit(X_train, y_train)
    y_pred = pipe.predict(X_test)
    acc = accuracy_score(y_test, y_pred)
    results[name] = acc
    print(f"{name} Accuracy: {acc:.4f}")
    print(classification_report(y_test, y_pred))
```

```python
for name, score in results.items():
    print(f"{name} Accuracy: {score:.4f}")
```

```python
import matplotlib.pyplot as plt
plt.bar(results.keys(), results.values(), color=['blue', 'red'])    
plt.ylabel('Accuracy Score')
plt.title('Comparison of Model Accuracy')   
plt.grid(True)
plt.show()

best_model_name = max(results, key=results.get)
best_model = models[best_model_name]
print(f"\n Best model: {best_model_name} with accuracy {results[best_model_name]:.4f}")
```
